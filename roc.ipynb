```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier

# Load data
df = pd.read_csv(r"C:\Users\alisa\OneDrive\Desktop\Alzamir\Alzheimer-s-Disease\alzheimers_disease_data.csv")
df.drop(columns=["PatientID", "DoctorInCharge"], inplace=True)
X = df.drop(columns=["Diagnosis"])
y = df["Diagnosis"]

# Feature selection
drop_features = ['Gender', 'Ethnicity', 'EducationLevel', 'Smoking', 'AlcoholConsumption',
                 'FamilyHistoryAlzheimers', 'CardiovascularDisease', 'Diabetes', 'Depression',
                 'SystolicBP', 'DiastolicBP', 'Confusion', 'Disorientation', 'DifficultyCompletingTasks',
                 'Forgetfulness']
X_fs = X.drop(columns=drop_features)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_fs_train, X_fs_test, y_fs_train, y_fs_test = train_test_split(X_fs, y, test_size=0.3, random_state=42)

# Scale for models that need it
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_fs_train_scaled = scaler.fit_transform(X_fs_train)
X_fs_test_scaled = scaler.transform(X_fs_test)

# Hyperparameters
best_params = {
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=200, learning_rate=0.1, max_depth=5,
                                                    min_samples_split=5, min_samples_leaf=2, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=200, max_depth=10, min_samples_split=5,
                                            min_samples_leaf=2, random_state=42),
    "XGBoost": XGBClassifier(n_estimators=300, learning_rate=0.05, max_depth=5, subsample=0.8,
                             colsample_bytree=0.8, random_state=42, use_label_encoder=False, eval_metric='logloss'),
    "CatBoost": CatBoostClassifier(depth=6, learning_rate=0.05, iterations=300, l2_leaf_reg=3,
                                   bagging_temperature=0.5, random_strength=5, verbose=0, random_state=42)
}

# Models
model_types = {
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'),
    "CatBoost": CatBoostClassifier(verbose=0, random_state=42)
}

# Evaluate all models
for model_name in model_types:
    print(f"\n================== {model_name} ==================")
    base_model = model_types[model_name]
    tuned_model = best_params[model_name]

    combos = [
        ("No FS, No Tuning", base_model, X_train_scaled, X_test_scaled, y_train, y_test),
        ("With FS Only", base_model, X_fs_train_scaled, X_fs_test_scaled, y_fs_train, y_fs_test),
        ("With Tuning Only", tuned_model, X_train_scaled, X_test_scaled, y_train, y_test),
        ("FS + Tuning", tuned_model, X_fs_train_scaled, X_fs_test_scaled, y_fs_train, y_fs_test),
    ] if model_name != "CatBoost" else [
        ("No FS, No Tuning", base_model, X_train, X_test, y_train, y_test),
        ("With FS Only", base_model, X_fs_train, X_fs_test, y_fs_train, y_fs_test),
        ("With Tuning Only", tuned_model, X_train, X_test, y_train, y_test),
        ("FS + Tuning", tuned_model, X_fs_train, X_fs_test, y_fs_train, y_fs_test),
    ]

    for label, model, xtr, xts, ytr, yts in combos:
        print(f"\nðŸ”¹ {label}")
        model.fit(xtr, ytr)
        preds = model.predict(xts)

        print("Classification Report:")
        print(classification_report(yts, preds))

        cm = confusion_matrix(yts, preds)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f"{model_name} - {label} Confusion Matrix")
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.tight_layout()
        plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# ROC plotting
for model_name in model_types:
    print(f"\nROC Curve for {model_name}")
    base_model = model_types[model_name]
    tuned_model = best_params[model_name]

    combos = [
        ("No FS, No Tuning", base_model, X_train_scaled, X_test_scaled, y_train, y_test),
        ("With FS Only", base_model, X_fs_train_scaled, X_fs_test_scaled, y_fs_train, y_fs_test),
        ("With Tuning Only", tuned_model, X_train_scaled, X_test_scaled, y_train, y_test),
        ("FS + Tuning", tuned_model, X_fs_train_scaled, X_fs_test_scaled, y_fs_train, y_fs_test),
    ] if model_name != "CatBoost" else [
        ("No FS, No Tuning", base_model, X_train, X_test, y_train, y_test),
        ("With FS Only", base_model, X_fs_train, X_fs_test, y_fs_train, y_fs_test),
        ("With Tuning Only", tuned_model, X_train, X_test, y_train, y_test),
        ("FS + Tuning", tuned_model, X_fs_train, X_fs_test, y_fs_train, y_fs_test),
    ]

    plt.figure(figsize=(8, 6))
    plt.title(f"{model_name} - ROC Curve Comparison")

    for label, model, xtr, xts, ytr, yts in combos:
        model.fit(xtr, ytr)
        try:
            y_score = model.predict_proba(xts)[:, 1]
        except AttributeError:
            try:
                y_score = model.decision_function(xts)
                y_score = (y_score - y_score.min()) / (y_score.max() - y_score.min())
            except:
                continue

        fpr, tpr, _ = roc_curve(yts, y_score)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"{label}")

    plt.plot([0, 1], [0, 1], 'k--', lw=1)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
